```python
pageviews = (traffic_data['event'] == 'pageview').sum()
previews = (traffic_data['event'] == 'preview').sum()
clicks = (traffic_data['event'] == 'click').sum()

print(pageviews)
print(previews)
print(clicks)
```

```python
preview_rate = previews/pageviews*100
print(preview_rate)
```

```python
conversion = clicks/(pageviews+previews)*100
print(conversion)
```
